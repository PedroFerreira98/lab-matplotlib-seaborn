```python
# MEDIAN STEPS BY MONTH_ENCODED
# your code here
import matplotlib.pyplot as plt

#I had to increase figsize in order to have labels on frame

fb = fitbit.groupby(by=['Months_encoded','Months'])['Steps'].median()

fb.plot(kind='bar',figsize=(12,10))

#plt.savefig('Months_Steps.png')


# MEDIAN STEPS BY WORK_OR_WEEKEND
# your code here

fb = fitbit.groupby(by=['Work_or_Weekend'])['Steps'].median()

fb.rename(index={0:'Weekend',1:'Workday'},inplace=True)

ax = fb.plot(kind='bar',figsize=(12,8),ylim=(0,11000),title='Steps Work vs Weekend')
for p in ax.patches:
    ax.annotate(np.round(p.get_height(),decimals=2),(p.get_x()+p.get_width()/2., p.get_height()), ha='center', va='center',xytext=(0, 10),textcoords='offset points')  
plt.xticks(rotation='horizontal')

#plt.savefig('Work_or_Weekend.png')
```

#### Plot the correlation matrix using seaborn.

```python
# your code here

corr = titanic.corr()
mask = np.triu(np.ones_like(corr, dtype=bool))

fig, ax = plt.subplots(figsize=(12, 10))

cmap = sns.diverging_palette(20, 230, as_cmap=True)

sns.heatmap(corr, mask=mask, cmap=cmap, vmax=1, center=0,vmin=-1,
            square=True, linewidths=.5, cbar_kws={"shrink": .5},annot=True)
```
